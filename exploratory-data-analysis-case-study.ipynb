```python
from urllib.request import urlretrieve  
urlretrieve("https://raw.githubusercontent.com/JovianML/opendatasets/master/data/stackoverflow-developer-survey-2020/survey_results_public.csv", "survey.csv")
```

```python
import pandas as pd
survey_df = pd.read_csv("survey.csv")
survey_df
```

```python
survey_df.info()
```

```python
survey_df.columns
```

```python
survey_df.shape
```

```python
survey_df.notnull()
```

```python
#fill the null values in column MainBranch with the value cs student 
survey_df["MainBranch"].fillna("cs student", inplace= False)
survey_df
```

```python
survey_df
```

```python
#fill the null values automatically
#(for numerical values)
#uses the previous values for the same
survey_df.interpolate()
```

```python
#adding column "language" with the value cpp
survey_df['language']=('cpp')
```

```python
#removing columns
survey_df.drop(columns='language')
```

```python
survey_df.Country
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
#ranking out some of the countries for plotting purposes
top_countries = survey_df.Country.value_counts().head(10)
top_countries
```

```python
plt.figure(figsize=(15,9))
sns.barplot(x=top_countries.index,y=top_countries)
```

```python
plt.figure(figsize=(12,6))
plt.hist(survey_df.Age)
```

```python
#tick refers to as the content on xaxis and y axis 
#you can refer to  the axis attribute for a particular axis
plt.figure(figsize=(20,12))
plt.tick_params(direction='out', length=13, width=5)

plt.xticks(fontweight='bold', fontsize=12)
plt.yticks(fontweight='bold', fontsize=12)
sns.scatterplot(x=survey_df.YearsCode, y=survey_df.YearsCodePro)
```
